```python
# TransUNet for Medical Image Segmentation with Performance Evaluation
# Complete implementation for Google Colab

# Step 1: Install required packages
!pip install torch torchvision torchaudio
!pip install timm
!pip install einops
!pip install opencv-python
!pip install scikit-image
!pip install matplotlib
!pip install numpy
!pip install Pillow
!pip install scipy

import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import os
from sklearn.model_selection import train_test_split
import timm
from einops import rearrange
from scipy.spatial.distance import directed_hausdorff
import warnings
warnings.filterwarnings('ignore')

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Step 2: TransUNet Implementation
class PatchEmbed(nn.Module):
    def __init__(self, img_size=224, patch_size=16, in_chans=3, embed_dim=768):
        super().__init__()
        self.img_size = img_size
        self.patch_size = patch_size
        self.n_patches = (img_size // patch_size) ** 2

        self.proj = nn.Conv2d(in_chans, embed_dim, kernel_size=patch_size, stride=patch_size)

    def forward(self, x):
        x = self.proj(x)  # (B, embed_dim, H//patch_size, W//patch_size)
        x = x.flatten(2).transpose(1, 2)  # (B, n_patches, embed_dim)
        return x

class MultiHeadAttention(nn.Module):
    def __init__(self, embed_dim, num_heads):
        super().__init__()
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.head_dim = embed_dim // num_heads

        self.qkv = nn.Linear(embed_dim, embed_dim * 3)
        self.proj = nn.Linear(embed_dim, embed_dim)

    def forward(self, x):
        B, N, C = x.shape
        qkv = self.qkv(x).reshape(B, N, 3, self.num_heads, self.head_dim).permute(2, 0, 3, 1, 4)
        q, k, v = qkv[0], qkv[1], qkv[2]

        attn = (q @ k.transpose(-2, -1)) / (self.head_dim ** 0.5)
        attn = attn.softmax(dim=-1)

        x = (attn @ v).transpose(1, 2).reshape(B, N, C)
        x = self.proj(x)
        return x

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, mlp_ratio=4.0):
        super().__init__()
        self.norm1 = nn.LayerNorm(embed_dim)
        self.attn = MultiHeadAttention(embed_dim, num_heads)
        self.norm2 = nn.LayerNorm(embed_dim)

        mlp_hidden_dim = int(embed_dim * mlp_ratio)
        self.mlp = nn.Sequential(
            nn.Linear(embed_dim, mlp_hidden_dim),
            nn.GELU(),
            nn.Linear(mlp_hidden_dim, embed_dim)
        )

    def forward(self, x):
        x = x + self.attn(self.norm1(x))
        x = x + self.mlp(self.norm2(x))
        return x

class VisionTransformer(nn.Module):
    def __init__(self, img_size=224, patch_size=16, in_chans=3, embed_dim=768,
                 depth=12, num_heads=12):
        super().__init__()
        self.patch_embed = PatchEmbed(img_size, patch_size, in_chans, embed_dim)
        self.pos_embed = nn.Parameter(torch.zeros(1, self.patch_embed.n_patches + 1, embed_dim))
        self.cls_token = nn.Parameter(torch.zeros(1, 1, embed_dim))

        self.blocks = nn.ModuleList([
            TransformerBlock(embed_dim, num_heads) for _ in range(depth)
        ])

        self.norm = nn.LayerNorm(embed_dim)

    def forward(self, x):
        B = x.shape[0]
        x = self.patch_embed(x)

        cls_tokens = self.cls_token.expand(B, -1, -1)
        x = torch.cat((cls_tokens, x), dim=1)
        x = x + self.pos_embed

        features = []
        for i, blk in enumerate(self.blocks):
            x = blk(x)
            if i in [2, 5, 8, 11]:  # Extract features at different depths
                features.append(x)

        return features

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, out_channels, skip_channels=0):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1)
        self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.bn2 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x, skip=None):
        if skip is not None:
            x = torch.cat([x, skip], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class TransUNet(nn.Module):
    def __init__(self, img_size=224, num_classes=1, embed_dim=768):
        super().__init__()
        self.img_size = img_size
        self.patch_size = 16
        self.embed_dim = embed_dim

        # Encoder (Vision Transformer)
        self.transformer = VisionTransformer(img_size, self.patch_size, 3, embed_dim)

        # CNN Encoder for skip connections
        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True)
        )

        self.conv2 = nn.Sequential(
            nn.MaxPool2d(2),
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True)
        )

        self.conv3 = nn.Sequential(
            nn.MaxPool2d(2),
            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True)
        )

        # Decoder
        self.decoder1 = DecoderBlock(embed_dim, 256, 256)
        self.decoder2 = DecoderBlock(256, 128, 128)
        self.decoder3 = DecoderBlock(128, 64, 64)
        self.decoder4 = DecoderBlock(64, 32)

        self.final_conv = nn.Conv2d(32, num_classes, 1)

    def forward(self, x):
        B = x.shape[0]

        # CNN encoder for skip connections
        skip1 = self.conv1(x)  # 224x224x64
        skip2 = self.conv2(skip1)  # 112x112x128
        skip3 = self.conv3(skip2)  # 56x56x256

        # Transformer encoder
        transformer_features = self.transformer(x)

        # Reshape transformer output
        patch_dim = self.img_size // self.patch_size
        x = transformer_features[-1][:, 1:]  # Remove CLS token
        x = x.transpose(1, 2).reshape(B, self.embed_dim, patch_dim, patch_dim)

        # Decoder with skip connections
        x = F.interpolate(x, size=(56, 56), mode='bilinear', align_corners=False)
        x = self.decoder1(x, skip3)

        x = F.interpolate(x, size=(112, 112), mode='bilinear', align_corners=False)
        x = self.decoder2(x, skip2)

        x = F.interpolate(x, size=(224, 224), mode='bilinear', align_corners=False)
        x = self.decoder3(x, skip1)

        x = self.decoder4(x)
        x = self.final_conv(x)

        return torch.sigmoid(x)

# Step 3: Dataset Class
class MedicalDataset(Dataset):
    def __init__(self, image_paths, mask_paths, transform=None, img_size=224):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.transform = transform
        self.img_size = img_size

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Load image
        image = cv2.imread(self.image_paths[idx])
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = cv2.resize(image, (self.img_size, self.img_size))

        # Load mask
        mask = cv2.imread(self.mask_paths[idx], cv2.IMREAD_GRAYSCALE)
        mask = cv2.resize(mask, (self.img_size, self.img_size))
        mask = (mask > 127).astype(np.float32)  # Binary threshold

        if self.transform:
            image = self.transform(image)
        else:
            image = torch.from_numpy(image.transpose(2, 0, 1)).float() / 255.0

        mask = torch.from_numpy(mask).unsqueeze(0).float()

        return image, mask

# Step 4: Evaluation Metrics
class MetricsCalculator:
    @staticmethod
    def dice_score(pred, target, smooth=1e-6):
        pred = (pred > 0.5).float()
        intersection = (pred * target).sum()
        dice = (2. * intersection + smooth) / (pred.sum() + target.sum() + smooth)
        return dice.item()

    @staticmethod
    def iou_score(pred, target, smooth=1e-6):
        pred = (pred > 0.5).float()
        intersection = (pred * target).sum()
        union = pred.sum() + target.sum() - intersection
        iou = (intersection + smooth) / (union + smooth)
        return iou.item()

    @staticmethod
    def hausdorff_distance(pred, target):
        pred_np = (pred.cpu().numpy() > 0.5).astype(np.uint8)
        target_np = target.cpu().numpy().astype(np.uint8)

        # Get contours
        pred_contours, _ = cv2.findContours(pred_np[0], cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        target_contours, _ = cv2.findContours(target_np[0], cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        if len(pred_contours) == 0 or len(target_contours) == 0:
            return float('inf')

        # Get boundary points
        pred_points = np.vstack(pred_contours).squeeze()
        target_points = np.vstack(target_contours).squeeze()

        if pred_points.ndim == 1:
            pred_points = pred_points.reshape(1, -1)
        if target_points.ndim == 1:
            target_points = target_points.reshape(1, -1)

        # Calculate Hausdorff distance
        hd1 = directed_hausdorff(pred_points, target_points)[0]
        hd2 = directed_hausdorff(target_points, pred_points)[0]

        return max(hd1, hd2)

# Step 5: Training Function
class Trainer:
    def __init__(self, model, device):
        self.model = model.to(device)
        self.device = device
        self.criterion = nn.BCELoss()
        self.optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
        self.metrics_calc = MetricsCalculator()

    def train_epoch(self, dataloader):
        self.model.train()
        total_loss = 0

        for images, masks in dataloader:
            images, masks = images.to(self.device), masks.to(self.device)

            self.optimizer.zero_grad()
            outputs = self.model(images)
            loss = self.criterion(outputs, masks)
            loss.backward()
            self.optimizer.step()

            total_loss += loss.item()

        return total_loss / len(dataloader)

    def evaluate(self, dataloader):
        self.model.eval()
        dice_scores = []
        iou_scores = []
        hd_scores = []

        with torch.no_grad():
            for images, masks in dataloader:
                images, masks = images.to(self.device), masks.to(self.device)
                outputs = self.model(images)

                for i in range(outputs.shape[0]):
                    dice = self.metrics_calc.dice_score(outputs[i], masks[i])
                    iou = self.metrics_calc.iou_score(outputs[i], masks[i])
                    hd = self.metrics_calc.hausdorff_distance(outputs[i], masks[i])

                    dice_scores.append(dice)
                    iou_scores.append(iou)
                    if hd != float('inf'):
                        hd_scores.append(hd)

        return {
            'dice': np.mean(dice_scores),
            'iou': np.mean(iou_scores),
            'hausdorff': np.mean(hd_scores) if hd_scores else float('inf'),
            'dice_std': np.std(dice_scores),
            'iou_std': np.std(iou_scores),
            'hausdorff_std': np.std(hd_scores) if hd_scores else 0
        }

# Step 6: Sample Data Generation (for demonstration)
def create_sample_data():
    """Create sample medical images and masks for demonstration"""
    os.makedirs('sample_data/images', exist_ok=True)
    os.makedirs('sample_data/masks', exist_ok=True)

    image_paths = []
    mask_paths = []

    for i in range(6):  # Create 6 sample images
        # Create synthetic medical image (grayscale with some structure)
        img = np.random.randint(50, 200, (224, 224, 3), dtype=np.uint8)

        # Add some circular structures to simulate organs/lesions
        center_x, center_y = np.random.randint(50, 174, 2)
        radius = np.random.randint(20, 40)

        y, x = np.ogrid[:224, :224]
        mask_circle = (x - center_x) ** 2 + (y - center_y) ** 2 <= radius ** 2
        img[mask_circle] = img[mask_circle] * 0.7 + 50  # Darken the region

        # Create corresponding binary mask
        mask = np.zeros((224, 224), dtype=np.uint8)
        mask[mask_circle] = 255

        # Save files
        img_path = f'sample_data/images/image_{i:02d}.png'
        mask_path = f'sample_data/masks/mask_{i:02d}.png'

        cv2.imwrite(img_path, cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
        cv2.imwrite(mask_path, mask)

        image_paths.append(img_path)
        mask_paths.append(mask_path)

    return image_paths, mask_paths

# Step 7: Main Training and Evaluation Pipeline
def main():
    print("Creating sample dataset...")
    image_paths, mask_paths = create_sample_data()

    # Split data
    train_imgs, val_imgs, train_masks, val_masks = train_test_split(
        image_paths, mask_paths, test_size=0.33, random_state=42
    )

    print(f"Training samples: {len(train_imgs)}")
    print(f"Validation samples: {len(val_imgs)}")

    # Create datasets
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    train_dataset = MedicalDataset(train_imgs, train_masks, transform=transform)
    val_dataset = MedicalDataset(val_imgs, val_masks, transform=transform)

    train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=1, shuffle=False)

    # Initialize model and trainer
    print("\nInitializing TransUNet model...")
    model = TransUNet(img_size=224, num_classes=1, embed_dim=768)
    trainer = Trainer(model, device)

    print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")

    # Training loop
    print("\nStarting training...")
    num_epochs = 20
    best_dice = 0

    for epoch in range(num_epochs):
        # Train
        train_loss = trainer.train_epoch(train_loader)

        # Evaluate
        val_metrics = trainer.evaluate(val_loader)

        print(f"Epoch {epoch+1}/{num_epochs}:")
        print(f"  Train Loss: {train_loss:.4f}")
        print(f"  Val Dice: {val_metrics['dice']:.4f} ± {val_metrics['dice_std']:.4f}")
        print(f"  Val IoU: {val_metrics['iou']:.4f} ± {val_metrics['iou_std']:.4f}")
        print(f"  Val HD: {val_metrics['hausdorff']:.2f} ± {val_metrics['hausdorff_std']:.2f}")

        # Save best model
        if val_metrics['dice'] > best_dice:
            best_dice = val_metrics['dice']
            torch.save(model.state_dict(), 'best_transunet_model.pth')
            print(f"  New best model saved! Dice: {best_dice:.4f}")

        print()

    # Final evaluation
    print("=" * 50)
    print("FINAL EVALUATION RESULTS")
    print("=" * 50)

    final_metrics = trainer.evaluate(val_loader)
    print(f"Final Dice Score: {final_metrics['dice']:.4f} ± {final_metrics['dice_std']:.4f}")
    print(f"Final IoU Score: {final_metrics['iou']:.4f} ± {final_metrics['iou_std']:.4f}")
    print(f"Final Hausdorff Distance: {final_metrics['hausdorff']:.2f} ± {final_metrics['hausdorff_std']:.2f}")

    # Visualize some predictions
    visualize_predictions(model, val_loader, device)

    return model, final_metrics

def visualize_predictions(model, dataloader, device, num_samples=3):
    """Visualize model predictions"""
    model.eval()
    fig, axes = plt.subplots(num_samples, 3, figsize=(12, 4*num_samples))

    with torch.no_grad():
        for idx, (images, masks) in enumerate(dataloader):
            if idx >= num_samples:
                break

            images, masks = images.to(device), masks.to(device)
            outputs = model(images)

            # Convert to numpy for visualization
            image = images[0].cpu().permute(1, 2, 0).numpy()
            image = (image * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406]))
            image = np.clip(image, 0, 1)

            mask = masks[0, 0].cpu().numpy()
            pred = (outputs[0, 0].cpu().numpy() > 0.5).astype(np.float32)

            axes[idx, 0].imshow(image)
            axes[idx, 0].set_title('Input Image')
            axes[idx, 0].axis('off')

            axes[idx, 1].imshow(mask, cmap='gray')
            axes[idx, 1].set_title('Ground Truth')
            axes[idx, 1].axis('off')

            axes[idx, 2].imshow(pred, cmap='gray')
            axes[idx, 2].set_title('Prediction')
            axes[idx, 2].axis('off')

    plt.tight_layout()
    plt.show()

# Run the main pipeline
if __name__ == "__main__":
    model, metrics = main()
```
